```
import torch
import torchvision
from torchvision import transforms, datasets

train = datasets.MNIST('', train=True, download=True,
                       transform=transforms.Compose([
                           transforms.ToTensor()
                       ]))

test = datasets.MNIST('', train=False, download=True,
                       transform=transforms.Compose([
                           transforms.ToTensor()
                       ]))


trainset = torch.utils.data.DataLoader(train, batch_size=10, shuffle=True)
testset = torch.utils.data.DataLoader(test, batch_size=10, shuffle=False)
```

```
import torch.nn as nn
import torch.nn.functional as F
```

```
class Net(nn.Module): #We'll call this class net and this net will inhereit from the nn
    def __init__(self):
        super().__init__()

net = Net()
print(net)
```

```
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(28*28, 64)
        self.fc2 = nn.Linear(64, 64)
        self.fc3 = nn.Linear(64, 64)
        self.fc4 = nn.Linear(64, 10)

        
    def forward(self, x):
        x = F.relu(self.fc1(x)) #用三層relu來激活,一般來說不激活也可以執行,但激活才有效果
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = self.fc4(x)#輸出層,如果在這裡用線性比較沒有意義
        return F.log_softmax(x, dim=1) #soft_max我們真正想的是每個機率在上面顯示,dim=1類似所有機率相加=1讓他更平滑,可以想像成一個軸




net = Net()
print(net)
```

```
X = torch.randn((28,28))
X = X.view(-1,28*28) #需固定28*28的比例不然output輸出錯誤,設定1或-1基本上都沒差
```

```
output = net(X) 
```
