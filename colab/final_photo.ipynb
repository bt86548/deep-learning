```
# --coding:big5--

from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
from keras.models import load_model
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import os

train_dir = r'/content/drive/My Drive/test_classify'

train_datagen =  ImageDataGenerator(rescale=1./255)
train_generator = train_datagen.flow_from_directory(train_dir)

print('='*30)
print('訓練的分類：',train_generator.class_indices)
print('='*30)

labels = train_generator.class_indices

#將分類做成字典方便查詢
labels = dict((v,k) for k,v in labels.items())
print(labels)

# 載入模型
model = load_model(r'/content/model_car_0827.h5')

# 將圖片轉為待測數據
def read_image(img_path):
    try:
        img = image.load_img(img_path, target_size=(224,224))
    except Exception as e:
        print(img_path,e)

    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    return img


# 隨機輸入一個待測圖片
# filename = r"E:\google drive\iii_course\HOT\img_distinct\classify\test_classify\Honda_Civic\honda_civic_2014_206880_01_l.jpg"
path = r'/content/drive/My Drive/test_classify/Honda_CR-V'
count = 0
count_sum = 0
for i in os.listdir(path):
    filename=path+'/'+i
    plt.figure()
    im = Image.open(filename)
    im_list = np.asarray(im)
    plt.title("predict")
    plt.axis("off")
    plt.imshow(im_list)
    # plt.show()
    img = read_image(filename)
    pred = model.predict(img)[0]
    # print(pred)
    # print(pred)
    m = max(pred)
    # print([i for i, j in enumerate(pred) if j == m])
    cor = [i for i, j in enumerate(pred) if j == m]
    name_split = i.split('_')
    name = name_split[0]+'_'+name_split[1]
    if labels[cor[0]]=='Honda_CR-V':
        count += 1
    print('辨識結果:',labels[cor[0]])
    count_sum += 1
    plt.close('all')
print('acc:{}%'.format((count/count_sum)*100))
```
